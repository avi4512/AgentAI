```python
from openai import OpenAI

# Initialize client
client = OpenAI(api_key="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",base_url="https://openrouter.ai/api/v1")


def email_response_agent(received_email, instruction_prompt):

    system_prompt = """
    You are a smart professional email assistant.
    Generate clear, polite, and context-aware email replies.
    Do not repeat the original email.
    Keep it properly formatted.
    """

    user_prompt = f"""
    Instruction:
    {instruction_prompt}

    Received Email:
    {received_email}

    Generate the appropriate email reply.
    """

    response = client.chat.completions.create(
        model="gpt-4o-mini",
        messages=[
            {"role": "system", "content": system_prompt},
            {"role": "user", "content": user_prompt}
        ],
        temperature=0.6
    )

    return response.choices[0].message.content


if __name__ == "__main__":
    email = input("Paste the received email:\n")
    prompt = input("Enter how you want to respond:\n")

    reply = email_response_agent(email, prompt)

    print("\nGenerated Email Reply:\n")
    print(reply)
```
