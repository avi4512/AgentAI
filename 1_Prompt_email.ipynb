```python
import os
import smtplib
from email.message import EmailMessage
from dotenv import load_dotenv
from openai import OpenAI

# Load environment variables
load_dotenv()

# Initialize OpenAI client
client = OpenAI(api_key="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",base_url="https://openrouter.ai/api/v1/")

# Email credentials
EMAIL_ADDRESS = "avi.punyala@gmail.com"
EMAIL_PASSWORD = "nuuofywyeydhkebu"


def generate_response(prompt):
    """Generate AI response"""
    response = client.chat.completions.create(
        model="gpt-4o-mini",
        messages=[
            {"role": "system", "content": "You are a helpful AI assistant."},
            {"role": "user", "content": prompt}
        ]
    )

    return response.choices[0].message.content


def send_email(subject, content, to_email):
    """Send email with AI response"""
    msg = EmailMessage()
    msg["Subject"] = subject
    msg["From"] = EMAIL_ADDRESS
    msg["To"] = to_email
    msg.set_content(content)

    # Gmail SMTP
    with smtplib.SMTP_SSL("smtp.gmail.com", 465) as smtp:
        smtp.login(EMAIL_ADDRESS, EMAIL_PASSWORD)
        smtp.send_message(msg)


def ai_email_agent():
    prompt = input("Enter your prompt: ")
    recipient = input("Enter recipient email: ")

    print("\nGenerating response...\n")
    ai_reply = generate_response(prompt)

    print("Sending email...\n")
    send_email("AI Agent Response", ai_reply, recipient)

    print("✅ Email sent successfully!")


if __name__ == "__main__":
    ai_email_agent()
```
